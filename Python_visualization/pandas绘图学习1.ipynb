```python
import pandas as pd
import numpy as np
ts = pd.Series(np.random.randn(1000), index=pd.date_range('1/1/2000', periods=1000))#一千个点
ts = ts.cumsum()#返回累加值，波动小，看出走势
ts.plot()
```

```python
df = pd.DataFrame(np.random.randn(1000, 4), index=ts.index, columns=list('ABCD'))
df = df.cumsum()
plt.figure()#Creates a new figure.创建一个新图
df.plot()#DataFrame plotting accessor and method
```

```python
df3 = pd.DataFrame(np.random.randn(1000, 2), columns=['B', 'C']).cumsum()
df3['A'] = pd.Series(list(range(len(df))))
df3.plot(x='A', y='B')
#取出一列声称另一列作为index一起绘图
```
