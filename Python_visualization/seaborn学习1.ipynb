```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
np.random.seed(sum(map(ord, "aesthetics")))
```

```python
def sinplot(flip=1):
    x = np.linspace(0, 14, 100)
    for i in range(1, 7):
        plt.plot(x, np.sin(x + i * .5) * (7 - i) * flip)
```

```python
sinplot()
```

```python
import seaborn as sns#引入seaborn模块
sinplot()#seaborn默认浅灰色背景与白色网络线
```

```python
#不加网格
sns.set_style("white")#五种主题风格：darkgrid, whitegrid, dark, white, and ticks
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data);
```

```python
#加网格
sns.set_style("whitegrid")#五种主题风格：darkgrid, whitegrid, dark, white, and ticks
data = np.random.normal(size=(20, 6)) + np.arange(6) / 2
sns.boxplot(data=data);
```

```python
sns.set_style("dark")#没有网格的dark风格
```

```python
sinplot()
```

```python
sns.set_style("white")#白色无网格风格
```

```python
sinplot()
```

```python
sns.set_style("ticks")#上方，右方轴上加了刻度
sinplot()
```

```python
sinplot()
sns.despine()#删除右边和上面的边界
```

```python
f, ax = plt.subplots()
sns.violinplot(data=data)
sns.despine(offset=10, trim=True)#trim使坐标轴不覆盖所有，两边加起来偏移10
```

```python
sns.set_style("whitegrid")
sns.boxplot(data=data, palette="deep")
sns.despine(left=True)#在despine中设置那些轴被删除，这里是删除左边的
```

```python
#使用with语句中的axes_style（）函数临时设置绘图参数
with sns.axes_style("darkgrid"):
    plt.subplot(211)
    sinplot()
plt.subplot(212)
sinplot(-1)
```

```python
sns.axes_style()#查看当前设置
```

```python
sns.set_style("darkgrid", {"axes.facecolor": ".9"})#设置参数，通过上述参数可以设定，字典传入set_style中
sinplot()
```

```python
#图和坐标显示的大小
sns.set()#重置参数
```

```python
sns.set_context("paper")# relative size, are paper, notebook, talk, and poster. The notebook style is the default
plt.figure(figsize=(8, 6))
sinplot()
```

```python
sns.set_context("talk")
plt.figure(figsize=(8, 6))
sinplot()
```

```python
sns.set_context("poster")
plt.figure(figsize=(8, 6))
sinplot()
```

```python
#调用set_context（）来设置参数，并且可以通过提供参数值的字典来覆盖参数
sns.set_context("notebook", font_scale=1.5, rc={"lines.linewidth": 2.5})
sinplot()
```
