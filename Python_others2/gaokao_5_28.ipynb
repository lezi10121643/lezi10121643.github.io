```python
#-*- coding=utf-8 -*-
from urllib import request
import re
import os,glob
import xlwt
import ssl

book = xlwt.Workbook()
sheet = book.add_sheet('sheet', cell_overwrite_ok=True)
path = 'D:\\work'
os.chdir(path)
result11=[]
result21=[]
result31=[]
result41=[]
#context = ssl._create_unverified_context()
ssl._create_default_https_context = ssl._create_unverified_context#关闭证书验证
for i in range(0,144):
    html=request.urlopen('https://gaokao.chsi.com.cn/zzbm/mdgs/detail.action?oid=476742435&lx=1&start='+str(30*i))
    #print(html.read().decode('UTF-8'))#避免出现\x8d\x9a\xe5\xae等
    response=html.read().decode('UTF-8')
    #print(response)
    pat='<td>(.*?)</td>'
    result = re.compile(pat).findall(str(response))
    #print(result)
    for i in range(0,30):
        result11.append(result[4*i+4])
        result21.append(result[4*i+5])
        result31.append(result[4*i+6])
        result41.append(result[4*i+7])
#print(result11)

j = 0
for i in range(0,len(result11)):
    try:
        zhiwei = result11[i]
        wangzhi = result21[i]
        gongzi = result31[i]
        gongzuodidian = result41[i]

        sheet.write(i + 1, j, zhiwei)
        sheet.write(i + 1, j + 1, wangzhi)
        sheet.write(i + 1, j + 2, gongzi)
        sheet.write(i + 1, j + 3, gongzuodidian)

    except Exception as e:
        print('出现异常：' + str(e))
        continue

book.save('d:\\shujufenxishi.xls')
```
