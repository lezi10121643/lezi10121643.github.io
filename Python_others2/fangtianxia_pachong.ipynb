```python
def gethousedetail(url):
    info={}
    res=requests.get(url)
    soup=BeautifulSoup(res.text,'html.parser')
    info['标题']=soup.select('.title h1')[0].text.strip()#.text拿出文字（title下面的h1标签）
    info['总价']=soup.select('.zongjia1 .red20b')[0].text
    for dd in soup.select('dd'):
        if '：' in dd.text.strip():#要有冒号的，用中文的冒号，因为网页中是中文
            key,value=dd.text.strip().split('：')#根据冒号切分
            info[key]=value
    return info
```

```python
import requests

from bs4 import BeautifulSoup

res=requests.get('http://esf.sh.fang.com/')#res.text#显示内容
domain='http://esf.sh.fang.com'#主域名（注意：http://esf.sh.fang.com/的最后/要拿掉，不然拼出来的不对）
soup = BeautifulSoup(res.text,'html.parser')
#print(soup)
houseary=[]#放房屋信息
for house in soup.select('.houseList dl'):#定位到class等于houseList,后面是<dl ......
    #print(house.select('.title')[0].text.strip())#strip()清除换行
    url=domain+house.select('.title a')[0]['href']#title下面的a链接里面的href(这里抓到的是相对链接)要+主域名
    #print(url)
    houseary.append(gethousedetail(url))
```
