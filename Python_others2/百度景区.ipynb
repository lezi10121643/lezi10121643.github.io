```python
#爬取多页
import re
import requests

for i in range(0,10):
    response = requests.get("https://lvyou.baidu.com/longmenshiku/remark/?rn=15&pn="+str(15*i)+"&style=hot#remark-container")
    response.encoding="utf-8"
    response = response.text
    pat = 'class="ri-remarktxt">(.*?)</div>'
    res = re.findall(pat,response)
    for i in res:
        print(i)
```
