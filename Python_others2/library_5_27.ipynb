```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_excel("被借过的书.xlsx")
```

```python
df
```

```python
a=''.join(df['正题名'])
```

```python
a
```

```python
import jieba
```

```python
wordsall = {} #define return dic
seg_list = jieba.cut(a, cut_all=False)
rowlist = ' '.join(seg_list)
words = rowlist.split(' ')
for word in words:
    if word !=' ':
        if word in wordsall:
            wordsall[word]+=1
        else:
            wordsall[word] = 1
wordsall = sorted(wordsall.items(), key=lambda d:d[1], reverse = True)
for (word,cut) in wordsall: 
    print ('%s:' % word,cut)        
```
