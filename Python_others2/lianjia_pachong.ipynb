```python

#主要程序
import requests

from bs4 import BeautifulSoup
houseary=[]#放房屋信息
domain='http://sh.lianjia.com'
for i in range(1,50):#先来十条
    res=requests.get('http://sh.lianjia.com/ershoufang/d'+str(i))#res.text#显示内容
    soup = BeautifulSoup(res.text,'html.parser')
    #print(soup)
    for j in range(0,29):
        url1=soup.select('.prop-title a')[j]['href']
        url=domain+url1
        houseary.append(gethousedetail1(url))
```

```python

def gethousedetail1(url):
    info={}
    q=requests.get(url)
    soup=BeautifulSoup(q.text,'html.parser')
    #for dd in soup.select('td'):
    for dd in soup.select('.content li'):
        a=dd.get_text(strip=True)#崔老师推荐的去空格方法，比strip（）好用
        #print(a)
        #print(dd.text.strip())
        if '：' in a:#要有冒号的，用中文的冒号，因为网页中是中文
            
            key,value=a.split('：')#根据冒号切分
            #print(a.split('：'))
            info[key]=value

    for dd in soup.select('.content li'):
        #print(dd.text.strip())
        b=dd.get_text(strip=True)#崔老师推荐的去空格方法，比strip（）好用
        if '：' in b:#要有冒号的，用中文的冒号，因为网页中是中文
            #print(b)
            key,value=b.split('：')#根据冒号切分
            #print(b.split('：'))
            info[key]=value
    info['总价']=soup.select('.bold')[0].text.strip()
    return info
```
