```python
from urllib import request
import re,xlwt,datetime

html=request.urlopen("https://read.douban.com/provider/all").read()    #
wzgz="<a href=\"(.*?)\" class=\"provider-item\"><div class=\"col-media\"><div class=\"cm-left avatar\"><div class=\"avatar\"><img src=\"(.*?)\"/></div></div><div class=\"cm-body\"><div class=\"name\">(.*?)</div><div class=\"works-num\">(.*?) </div></div></div></a>"
xx=re.compile(wzgz).findall(str(html,"utf-8"))                         #
#print(xx)

#workbooksheet
workbook = xlwt.Workbook()
sheet1 = workbook.add_sheet('sheet1',cell_overwrite_ok=True)

#excel
style = xlwt.XFStyle()

#
font = xlwt.Font()
font.name = 'Times New Roman'
font.bold = True

#
style.font = font

#sheet11
sheet1.write(0,0,"",style)
sheet1.write(0,1,"-URL",style)
sheet1.write(0,2,"LOGO-URL",style)
sheet1.write(0,3,"",style)
sheet1.write(0,4,"",style)

a=0                                                                #
h=0                                                                #
for i in xx:
    #print(str(a+1),i[0])
    sheet1.write(a+1,0,a+1,style)                                   #a+11
    sheet1.write(a+1,1,"https://read.douban.com"+str(i[0]),style)   #a+12URL
    sheet1.write(a+1,2,i[1],style)                                  #a+13LOGO-URL
    sheet1.write(a+1,3,i[2],style)                                  #a+14
    sheet1.write(a+1,4,int(i[3]),style)                             #a+15
    h+=int(i[3])                                                    #
    a+=1

    if a==a:                                                        #XXsheet1
        t=datetime.datetime.now().strftime("%Y-%m-%d %H:%M:%S")
        t1=datetime.datetime.now().strftime("%Y%m%d%H%M%S")
        sheet1.write(a+1,3,"",style)                            #sheet1
        sheet1.write(a+1,4,h,style)                                 #sheet1
        sheet1.write(a+2,3,"",style)                        #sheet1
        sheet1.write(a+2,4,t,style)                                 #sheet1

workbook.save("d:/"+str(t1)+".xls")                 #excel,

print("excel")
print(""+str(h))
```
