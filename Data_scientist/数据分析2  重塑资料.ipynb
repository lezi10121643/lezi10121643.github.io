```python
import pandas
df = pandas.read_excel('house_sample.xlsx')
df.head()
```

```python
pandas.get_dummies(df['朝向'])#建立虚拟变量，是南则在南位置显示1，以此类推
```

```python
df = pandas.concat([df, pandas.get_dummies(df['朝向'])], axis = 1)#合并虚拟变量和原DataFrame
df.head()
```

```python
df = df.drop('朝向', axis = 1)#按列删除朝向字段，因为已经不需要了
df.head()
```

```python
?df.pivot_table#建立透视表
```

```python
df2 = df.pivot_table(index = '张贴日期', columns='产权性质', values='总价', aggfunc=sum, fill_value='0')
df2.head()#张贴日期作为索引，产权性质作为栏位，总价作为值
```

```python
df3 = df.pivot_table(index = '产权性质', columns='张贴日期', values='总价', aggfunc=sum, fill_value=0)
df3.head()
```
