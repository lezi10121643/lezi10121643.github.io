```python
def Fib(n):
    a,b = 0,1
    for i in range(n-1):
        a,b = b,a+b
    return b
```

```python
def Fib(n):
    if n == 0:
        return 0
    a,b = 0,1
    for i in range(n-1):
        a,b = b,a+b
    return b
```

```python
Fib(0)
```

```python
Fib(10)
```

```python
def Fib(n):
    if n == 0:
        yield 0
    a,b = 0,1
    for i in range(n-1):
        a,b = b,a+b
    yield b
```

```python
Fib(10)
```

```python
#看上去同时执行的就是多任务____协程     （除此之外还有线程，进程）
```

```python
def func1():
    while 1:
        print('A:装入货物')
        yield None
        
def func2():
    while 1:
        print('B：打包货物')
        yield None
```

```python
f1 = func1()
f2 = func2()
while 1:
    f1.__next__()
    f2.__next__()
```

```python
#协程，又称微线程，纤程。英文名Coroutine。
```
